```python
from nltk.util import ngrams
from collections import Counter
```

```python
df['kmers_3'] = df['sequence'].apply(
    lambda s: [''.join(k) for k in ngrams(s, 3)]
)
```

```python
df.head(10)
```

```python
df['sequence'] = df['sequence'].str.replace('T', 'U')
```

```python
df.head()
```

```python
import itertools

bases = ['A', 'C', 'G', 'U']
vocab = [''.join(p) for p in itertools.product(bases, repeat=3)]
```

```python
from collections import Counter
import numpy as np

def kmer_freq_vector(kmers, vocab):
    counts = Counter(kmers)
    return np.array([counts.get(k, 0) for k in vocab])
```

```python
X = np.vstack(
    df['kmers_3'].apply(lambda x: kmer_freq_vector(x, vocab))
)
```

```python
from sklearn.preprocessing import normalize

X_norm = normalize(X, norm='l2')
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cos_sim_matrix = cosine_similarity(X_norm)
```
